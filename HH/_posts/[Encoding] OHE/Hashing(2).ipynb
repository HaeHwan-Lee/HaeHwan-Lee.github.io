# HashingEncoder VS. BinaryEncoder

```python
import pandas as pd
    
train = pd.read_csv('https://raw.githubusercontent.com/kiwidamien/StackedTurtles/master/content/preprocessing/simple_loan_example.csv')
train
```

> train으로 사용할 데이터

```python
import numpy as np

test = pd.DataFrame(data=np.array([[13000, 0.45, 2500, 'refinance', 'A', 'True'], [74000, 0.15, 23000, 'NEW', 'C', 'False']]), columns = train.columns)
test
```

> test로 사용할 데이터 : 두번째 row의 경우 purpose에 새로운 값이 등장하고 있다.

## BinaryEncoding

```python
from category_encoders import *
```

```python
bina = BinaryEncoder(drop_invariant = True, cols = ['purpose'])
bina.fit_transform(train)
```

#### 새로운 테스트 데이터를 마주칠 경우

```python
bina.transform(test)
```

> 새로운 데이터를 모두 0벡터로 반환했다.

<hr>

## HashingEncoder

```python
ha = HashingEncoder(n_components=4, cols=['purpose'], max_process = 1)
ha.fit_transform(train)
```

#### 새로운 테스트 데이터를 마주칠 경우

```python
ha.transform(test)
```

> 새로운 데이터셋은 [1, 0, 0, 0]으로 새로운 값이 피팅되었음을 확인할 수 있다.
